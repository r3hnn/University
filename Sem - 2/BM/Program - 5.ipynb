```python
from sympy import*
x=Symbol('x')
y=Symbol('y')
z=Symbol('z')
```

```python
u=x**3+y**3+z**3-3 *x *y *z
S=simplify(x *diff(u,x)+y *diff(u,y)+z *diff(u,z))
E=3*u
if simplify(S-E)==0:
  print('Euler’s Theorem is Satisfied.')
else:
  print('Euler’s Theorem is not Satisfied.')
```

```python
u=x**4 * ln(y/x)
S=simplify(x *diff(u,x)+y *diff(u,y))
E=4*u
if simplify(S-E)==0:
  print('Euler’s Theorem is Satisfied.')
else:
  print('Euler’s Theorem is not Satisfied.')
```

```python
# New function: u = (x^5 + y^5) / (x + y)
u = (x**5 + y**5) / (x + y)

# Calculate the Left Hand Side (LHS) of Euler's equation: xdu/dx + ydu/dy
# We simplified the expression to make comparison easier
LHS = simplify(x *diff(u, x) + y *diff(u, y))

# The degree is 4, so the Right Hand Side (RHS) should be 4*u
n = 4
RHS = n * u

print(f"LHS (x *du/dx + y *du/dy) simplified: {LHS}")
print(f"RHS (n*u) simplified: {simplify(RHS)}")
print("-" * 30)

# Check if LHS - RHS is zero
if simplify(LHS - RHS) == 0:
    print(f"Euler’s Theorem is Satisfied for degree n = {n}.")
else:
    print("Euler’s Theorem is not Satisfied.")
```
