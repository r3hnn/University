```python
import numpy as np
from scipy.optimize import linprog
```

```python
c = [3, 1]
A = [
    [3, 4],
    [-4, -7],
    [0, 1]
]
b = [23, -34, 4]

x_bounds = (3, None)
y_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# 5. Print Result
if res.success:
    print(f"Optimal Value: Z = {res.fun}")
    print(f"Coordinates: x = {res.x[0]}, y = {res.x[1]}")
else:
    print("No solution found:", res.message)
```

```python
# Convert maximization to minimization
c = [-3, -5]

A = [
    [3, 2],   # 3x + 2y ≤ 18
    [1, 0],   # x ≤ 4
    [0, 1]    # y ≤ 6
]

b = [18, 4, 6]

# Bounds for x and y
x_bounds = (0, None)
y_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Print result
if res.success:
    print(f"Optimal Value: Z = {-res.fun}")
    print(f"Coordinates: x = {res.x[0]}, y = {res.x[1]}")
else:
    print("No solution found:", res.message)
```

```python
# Objective function (maximize → minimize negative)
c = [-5, -1, -4]

# Constraint matrix
A = [
    [0, 1, 1],
    [1, 1, 1],
    [1, 0, 1]
]

b = [3, 5, 8]

# Bounds (x, y, z ≥ 0)
bounds = [(0, None), (0, None), (0, None)]

# Solve
res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Output
if res.success:
    print(f"Optimal Value: Z = {-res.fun}")
    print(f"Coordinates: x = {res.x[0]}, y = {res.x[1]}, z = {res.x[2]}")
else:
    print("No solution found:", res.message)
```
