Find Jacobian Matrix
</h2>

```python
from sympy import *
x, y = symbols('x y')
u=x**2-2*y
v=x+y
a = diff(u, x)
b = diff(v, x)
c = diff(u, y)
d = diff(v, y)
M = Matrix([[a, b], [c, d]])
pprint(M)
print('Jacobian =', M.det())
```

```python
import numpy as np
import matplotlib.pyplot as plt

def transform(x, y):
    u = x + y
    v = x * y
    return u, v

fig, ax = plt.subplots()

# Create grid lines in (x,y)
t = np.linspace(-3, 3, 200)

# Lines x = constant
for x0 in np.linspace(-3, 3, 10):
    y = t
    x = np.full_like(t, x0)
    u, v = transform(x, y)
    ax.plot(u, v)

# Lines y = constant
for y0 in np.linspace(-3, 3, 10):
    x = t
    y = np.full_like(t, y0)
    u, v = transform(x, y)
    ax.plot(u, v)

ax.axhline(0)
ax.axvline(0)
ax.set_title("Transformation (u = x+y, v = xy)")
ax.grid(True)

plt.show()
```
