```r
num1 <- as.integer(readline(prompt = "Enter 1st number: "))
num2 <- as.integer(readline(prompt = "Enter 2nd number: "))
```

```r
add <- num1 + num2
sub <- num1 - num2
mul <- num1 * num2
div <- num1 / num2
mod_div <- num1 %% num2
```

```r
cat("The sum is: ", add, "\n")
cat("The difference is: ", sub, "\n")
cat("The product is: ", mul, "\n")
cat("The quotient is: ", div, "\n")
cat("The remainder is: ", mod_div, "\n")
```

```r
sales <- matrix(
  c(120,150,130,160,
    100,140,120,150,
    90,110,100,130),
  nrow=3,
  byrow=TRUE)
```

```r
price <- matrix(
  c(10,10,10,10,
    12,12,12,12,
    8,8,8,8),
  nrow=3,
  byrow=TRUE)
```

```r
revenue <- sales * price
revenue
```

```r
par(mar = c(1, 1, 1, 1))
plot(1:10,
     xlab = "THIS IS A VERY VERY VERY LONG X LABEL",
     ylab = "THIS IS A VERY VERY VERY LONG Y LABEL",
     main = "A VERY LONG TITLE TO BREAK THE PLOT")
```

```r
region_revenue <- rowSums(revenue)
region_revenue
```

```r
image <- matrix(
  c(50,60,70,80,
    90,100,110,120,
    130,140,150,160,
    170,180,190,200),
  nrow=4,
  byrow=TRUE)

image
```

```r
bright <- image + 20
bright
```

```r
filter <- matrix(0.5, nrow=4, ncol=4)
filtered <- bright * filter
filtered
```

```r
normalized <- filtered / max(filtered)
normalized
```

```r
rainfall <- matrix(sample(1:100, 50, replace=TRUE), nrow=5)
rainfall
```

```r
avg_rainfall <- rowMeans(rainfall)
avg_rainfall
```

```r
which.max(avg_rainfall)
```

```r
contact <- matrix(
  c(5,3,1,1,
    3,6,2,1,
    1,2,4,2,
    1,1,2,3),
  nrow=4,
  byrow=TRUE)

contact
```

```r
infected <- matrix(c(50, 80, 20, 10), ncol=1)

infected
```

```r
exposure <- contact %*% infected
exposure
```

```r
which.max(exposure)
```
