</h2 style = "text-align:center;>
Round Robin Scheduling
</h2>

```python
n = int(input("Enter the number of processes: "))
```

```python
at = []
bt = []
```

```python
for i in range(n):
    at.append(int(input(f"Enter arrival time for P{i + 1}: ")))
    bt.append(int(input(f"Enter burst time for P{i + 1}: ")))
```

```python
tq = int(input("Enter time quantum: "))
```

```python
rt = bt.copy()
ct = [0] * n
time = 0
completed = 0
queue = []
visited = [False] * n
```

```python
while completed < n:
	for i in range(n):
		if at[i] <= time and not visited[i]:
			queue.append(i)
			visited[i] = True

	if not queue:
		time += 1
		continue

	i = queue.pop(0)

	if rt[i] > tq:
		time += tq
		rt[i] -= tq
	else:
		time += rt[i]
		rt[i] = 0
		ct[i] = time
		completed += 1

	for j in range(n):
		if at[j] <= time and not visited[j]:
			queue.append(j)
			visited[j] = True

	if rt[i] > 0:
		queue.append(i)
```

```python
tat = [0] * n
wt = [0] * n
```

```python
for i in range(n):
    tat[i] = ct[i] - at[i]
    wt[i] = tat[i] - bt[i]
```

```python
print("\nProcess\tAT\tBT\tWT\tTAT\tCT")
for i in range(n):
	print(f"P{i + 1}\t{at[i]}\t{bt[i]}\t{wt[i]}\t{tat[i]}\t{ct[i]}")
```

```python
print("\nAverage Waiting Time:", sum(wt) / n)
print("Average Turnaround Time:", sum(tat) / n)
```
