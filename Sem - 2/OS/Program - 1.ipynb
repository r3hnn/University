```python
n = int(input("Enter the number of processes: "))
```

```python
burst_time = []
for i in range(n):
    bt = int(input("Enter the burst time: "))
    burst_time.append(bt)
```

```python
waiting_time = [0] * n
turnaround_time = [0] * n
```

```python
for i in range(1, n):
    waiting_time[i] = waiting_time[i - 1] + burst_time[i - 1]
```

```python
for i in range(n):
    turnaround_time[i] = waiting_time[i] + burst_time[i]
```

```python
print("Process\tBT\tWT\tTAT")
for i in range(n):
    print(f"P{i + 1}\t{burst_time[i]}\t{waiting_time[i]}\t{turnaround_time[i]}")
```

```python
print("\nThe average waiting time is: ", sum(waiting_time) / n)
print("\nThe average turnaround time is: ", sum(turnaround_time) / n)
```

```python
import matplotlib.pyplot as plt
t = 0
plt.figure(figsize=(10, 2))

for i, bt in enumerate(burst_time):
    plt.barh("CPU", bt, left=t)
    plt.text(t + bt / 2, 0, f"P{i + 1}", ha='center', va='center')
    plt.text(t , -0.15,t, ha='center')
    t+= bt
    plt.text(t, -0.15, t, ha='center')

plt.xlabel("Time")
plt.title("First Come First Serve (FCFS) Scheduling")
plt.show()
```
