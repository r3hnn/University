<h2 style = "text-align:center;">
Shortest Seek Time First (SSTF) Disk Scheduling
</h2>

```python
head = 50
seek_time = list(map(int, input("Enter the seek time for each disk (Space Separated): ").split()))
```

```python
total_seek = 0
step = 1
```

```python
print("Step\tHead\t\tNext Seek\tSeek Distance")
while seek_time:
    distance = [abs(r - head) for r in seek_time]
    idx = distance.index(min(distance))

    next_seek = seek_time[idx]
    seek = abs(next_seek - head)

    print(f"{step}\t{head}\t\t{next_seek}\t\t{seek}")
    total_seek += seek
    head = next_seek
    seek_time.pop(idx)
    step += 1

print(f"\nTotal Seek Distance: {total_seek}")
```
