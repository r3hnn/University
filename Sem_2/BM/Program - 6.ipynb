Find Jacobian Matrix
</h2>

```python
from sympy import *
x, y = symbols('x y')
```

```python
u = 3*x + 5*y
v = 4*x-3*y
a = diff(u, x)
b = diff(u, y)
c = diff(v, x)
d = diff(v, y)
M = Matrix([[a, b], [c, d]])
print(M)
print('Jacobian =', M.det())
```
