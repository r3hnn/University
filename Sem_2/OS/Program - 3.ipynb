<h2 style = "text-align:center;">
Priority Scheduling (Lower number = Higher Priority)
</h2>

```python
n = int(input("Enter the number of processes: "))
processes = []
for i in range(n):
    bt = int(input(f"Enter the burst time: "))
    pr = int(input(f"Enter the priority: "))
    processes.append((i+1,bt, pr))
```

```python
processes.sort(key=lambda x: x[2])
waiting_time = [0]*n
for i in range(1, n):
    waiting_time[i] = waiting_time[i-1] + processes[i-1][1]
print("Process\tBT\tPriority\tWT")
for i in range(n):
    print(f"P{processes[i][0]}\t{processes[i][1]}\t{processes[i][2]}\t\t{waiting_time[i]}")
print("The average waiting time is: ", sum(waiting_time)/n)
```

```python
import matplotlib.pyplot as plt
process_id = [f"P{p[0]}" for p in processes]
bars = plt.bar(process_id, waiting_time)

for i in range(len(waiting_time)):
    plt.text(i, waiting_time[i], str(waiting_time[i]), ha='center', va='bottom')

plt.xlabel("Process")
plt.ylabel("Waiting Time")
plt.title("Waiting Time of Processes (Priority Scheduling)")
plt.show()
```
