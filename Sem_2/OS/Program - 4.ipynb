<h2 style = "text-align:center;">
Shortest Remaining Time First
</h2>

```python
n = int(input("Enter the Number of Processes: "))
```

```python
pid = []
at = []
bt = []
rt = []
```

```python
for i in range(n):
    pid.append(i+1)
    at.append(int(input(f"Enter the Arrival Time for P{i+1}: ")))
    bt.append(int(input(f"Enter the Burst Time for P{i+1}: ")))
    rt.append(bt[i])
```

```python
ct = [0] * n
wt = [0] * n
tat = [0] * n
```

```python
time = 0
completed = 0
```

```python
while completed < n:
	idx = -1
	min_bt = 9999
	for i in range(n):
		if at[i] <= time and rt[i] > 0:
			if rt[i] < min_bt or (rt[i] == min_bt and at[i] < at[idx]):
				min_bt = rt[i]
				idx = i
	if idx == -1:
		time += 1
		continue
	rt[idx] -= 1
	time += 1
	if rt[idx] == 0:
		ct[idx] = time
		tat[idx] = ct[idx] - at[idx]
		wt[idx] = tat[idx] - bt[idx]
		completed += 1
```

```python
print("\nPID\tAT\tBT\tWT\tTAT\tCT")
for i in range(n):
	print(f"{pid[i]}\t{at[i]}\t{bt[i]}\t{wt[i]}\t{tat[i]}\t{ct[i]}")
```

```python
print("\nAverage Waiting Time:", sum(wt) / n)
print("Average Turnaround Time:", sum(tat) / n)
```
