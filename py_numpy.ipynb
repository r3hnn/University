```python
# Python list
my_list = [1, 2, 3, 4, 5]

# NumPy array
my_array = np.array([1, 2, 3, 4, 5])
```

```python
# Python list — doubles the list (repeats it)
print(my_list * 2)

# NumPy array — doubles every value
print(my_array * 2)
```

```python
salaries = np.array([50000, 60000, 70000, 55000])

# ❌ Without vectorization — slow loop
for i in range(len(salaries)):
    salaries[i] = salaries[i] * 1.10

# ✅ With vectorization — one line, much faster
salaries = salaries * 1.10
salaries
```

```python
temperatures = [36.5, 38.2, 37.0, 39.1, 36.8]
for i in range(len(temperatures)):
	if temperatures[i] > 38.0:
		print("fever")
	else:
		print("normal")
```

```python
scores = [45, 82, 61, 39, 75, 90, 55]
for i in scores:
    if i >= 50:
        print("Pass")
    else:
        print("Fail")
```

```python
count = 0
for score in scores:
    if score >= 50:
        print(f"{score} → Pass")
        count = count + 1  # add 1 every time we pass
    else:
        print(f"{score} → Fail")

print(f"{count} students passed")
```

```python
products = [
	{"name": "Laptop", "price": 999},
	{"name": "Phone",  "price": 699},
	{"name": "Tablet", "price": 449},
]

for product in products:
    if product["price"] > 600:
        print({product['name']})
```

```python
def min_max(data):
    return (min(data), max(data))   # return two values at once

low, high = min_max([4, 7, 2, 9, 1])
print(low)    # → 1
print(high)   # → 9
```
