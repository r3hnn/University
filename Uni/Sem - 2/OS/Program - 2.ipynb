Shortest Job First - non-Premptive Scheduling

```python
n = int(input("Enter the number of Processess: "))
processes = []
for i in range(n):
    at = int(input("Enter the Arrival Time of Process: "))
    bt = int(input("Enter the Burst Time of Process: "))
    processes.append([i+1, at, bt])
```

```python
time = 0
completed = []
wt_list = []
tat_list = []
print("\nProcess\tAT\tBT\tWT\tTAT\tCT")

while len(completed) < n:
    available = [p for p in processes if p[1] <= time and p not in completed]
    if not available:
        time += 1
        continue
    job = min(available, key = lambda x:x[2])
    pid, at, bt = job
    wt = time - at
    ct = time + bt
    tat = ct - at
    wt_list.append(wt)
    tat_list.append(tat)
    print(f"P{pid}\t{at}\t{bt}\t{wt}\t{tat}\t{ct}")
    time = ct
    completed.append(job)
```

```python
print("\nThe Average Waiting Time is: ", sum(wt_list)/n)
print("\nThe Average Turn Around Time is: ", sum(tat_list)/n)
```
