```python
from sympy import *
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
```

```python
print(diff(sin(4*x)*cos(3*x), x))
```

```python
u = x**3+y**3+z**3-3*x*y*z
s = simplify(x*diff(u,x)+y*diff(u,y)+z*diff(u,z))
s
```

```python
f2=(1-2*x*y+y**2)**Rational(-1,2)
ans=x*diff(f2,x)-y*diff(f2,y)
S=simplify(ans)
S
```

```python
f3=x**2*y+y**2*z+z**2*x
ans=diff(f3,x)+diff(f3,y)+diff(f3,z)
S=simplify(ans)
S
```

```python
u=x**2*y+y**2*z+x**2*y
ans=diff(u,x)+diff(u,y)+diff(u,z)
S=simplify(ans)
S
```

```python
u= x**y +y**x
ans=diff(u,x)+diff(u,y)
S=simplify(ans)
S
```

```python
u = ln(x**4+y**4*x-y)
ans=diff(u,x)+diff(u,y)+diff(u,z)
S=simplify(ans)
S
```

```python
u = x**2*y + y
expr = x**2*diff(u,x,2) + y**2*diff(u,y,2) + 2*x*y*diff(u,x,y)
S = simplify(expr)
S
```

```python
u = sqrt(x**2 + y**2 + z**2)
expr = diff(u,x)**2 + diff(u,y)**2 + diff(u,z)**2
S = simplify(expr)
S
```

```python
u = x**3 - 2*x**2*y + 3*x*y**2 + y**3
expr = x**2*diff(u,x) + y**2*diff(u,y)
S = simplify(expr)
S
```

```python
u = 3*x**2*y*z + 4*x*y**2*z + 5*y**4
expr = diff(u,x) + diff(u,y) + diff(u,z)
S = simplify(expr)
S
```

```python
u = exp(x)*(x*cos(y) - y*sin(y))
expr = diff(u,x,2) + diff(u,y,2)
S = simplify(expr)
S
```

```python
u = y/z + z/x + x/y
expr = x*diff(u,x) + y*diff(u,y) + z*diff(u,z)
S = simplify(expr)
S
```

```python
u = (x*y)/(x+y)
expr = x**2*diff(u,x,2) + y**2*diff(u,y,2) + 2*x*y*diff(u,x,y)
S = simplify(expr)
S
```
